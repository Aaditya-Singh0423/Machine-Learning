```python
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Amit', 'Priya', 'Rahul', 'Sneha', 'Karan'],
    'Age': [23, 25, 22, 24, 26],
    'City': ['Delhi', 'Mumbai', 'Pune', 'Bangalore', 'Chennai'],
    'Score': [85, 90, 78, 88, 92]
}                                                                               # Sample data

df2 = pd.DataFrame(data, index=['s1', 's2', 's3', 's4', 's5'])
```

```python
df2
```

```python
df2.iloc[0:2,0]
```

```python
df2.iloc[0:2,[0,2]]                    #first 2 rows and first and third columns #fancy indexing
```

```python
df2.loc["s1":"s3"]                      #label based indexing on basis of name
```

```python
df2.loc["s1":"s3",["Name","City"]]          #label based indexing
```

```python
df2.columns
```

```python
#change the city name with age 25 tp kolkata using iloc
cond=df2['Age']==25
cond
```

```python
X=df2.loc[cond]
X
```

```python
X.loc[:,"Age"]=22
X
```

```python
df2.loc[df2["Age"]==25,"City"]="Kolkata"               #changing value using loc
df2
```

```python
df2.loc[(df2["Age"]>20)&(df2["Age"]<=25),"Age"]=22               #changing value using loc
df2
```
