```python
def clean_column(ele):
    ele=ele.strip()                           # Remove leading/trailing whitespace
    ele=ele.lower()                           # Convert to lowercase
    ele=ele.replace(" ","_")                  # Replace spaces with underscores
    ele=ele.replace("(","")                   # Remove opening parentheses
    ele=ele.replace(")","")                   # Remove closing parentheses
    ele=ele.replace("operating_system","os")  # Abbreviate 'operating_system' to 'os'
    ele=ele.replace("os_version","osv")       # Abbreviate 'Operating_System' to 'os'
    return ele
```

```python
for name in df.columns:
    clean_name=clean_column(name)
    df.rename(columns={name:clean_name},inplace=True)          # Rename columns using the clean_column function
```

```python
def convert_to_gb(value):                                            #convert storage to GB
    parts = re.findall(r'(\d+\.?\d*)\s*(TB|GB)', str(value))         #find the number and unit
    total_gb = 0                                                     #initialize total_gb
    for amount, unit in parts:                                       #iterate over found parts
        amount = float(amount)                                       #convert to float
        if unit.upper() == 'TB':                                     #convert TB to GB
            amount *= 1024                                           # Convert TB â†’ GB
        total_gb += amount                                           #sum up the total_gb
    return total_gb                                                  # Return total storage in GB
```

```python
df['storage_gb'] = df['storage'].apply(convert_to_gb).astype(int)    # Apply the conversion function and convert to integer
```
