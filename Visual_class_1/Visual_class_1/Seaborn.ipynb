```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
sns.relplot(data=df,x='total_bill',y='tip',hue='sex',col='smoker', style='time',size='size')
```

```python
sns.relplot(data=df,x='total_bill',y='tip',hue='sex',col='day', style='time',size='size', col_wrap=2, height=4, aspect=1)
plt.show()
```

```python
df.groupby(['day','size'])[['total_bill','tip']].mean().reset_index()
```

```python
sns.scatterplot(data=df,x='total_bill',y='tip')
plt.show()
```

```python
df.head()
```

```python
sns.displot(data=df,x='total_bill')
```

```python
sns.histplot(data=df,x='total_bill',bins=30,hue='time',multiple='stack')
```

```python
sns.histplot(data=df,x='total_bill',bins=30,hue='time',element="step")
```

```python
sns.kdeplot(data=df,x='total_bill',hue='time',fill=True)
```

```python
sns.histplot(data=df,x='total_bill',y="tip",hue="smoker",element="step",fill=False)
```

```python
sns.kdeplot(data=df,x='total_bill',y="tip",fill=True)
```

```python
m=df.pivot_table(index='day',columns='time',values='size',aggfunc='sum')
```
