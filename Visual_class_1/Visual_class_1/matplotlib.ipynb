```python
def Country_plot(country_name,linestyle,color):
    country_data=df[df['Country']==country_name]
    plt.plot(country_data['Date_reported'],country_data['Cumulative_cases'],ls=linestyle,color=color,label=country_name)              #line plot
    plt.xlabel('Date_reported')
    plt.ylabel('Cumulative_cases')
    plt.title('Covid Cases')
    plt.legend(loc="upper left")
    plt.show()
Country_plot('Italy','--','blue')
Country_plot('Spain','-.','green')
Country_plot('India',':','orange')
```

```python
def Country_graph(**kwargs):
    for country_name,style_color in kwargs.items():
        linestyle,color=style_color
        country_data=df[df['Country']==country_name]
        plt.plot(country_data['Date_reported'],country_data['Cumulative_cases'],ls=linestyle,color=color,label=country_name)              #line plot
    plt.xlabel('Date_reported')
    plt.ylabel('Cumulative_cases')
    plt.title('Covid Cases')
    plt.legend(loc="upper left")
    plt.show()
Country_graph(Italy=('--','blue'),Spain=('-.' ,'green'),India=(':','orange'),China=('--','red'))
```

```python
# create a clustered bar graph for casual and registered users for each season without using .plot(kind="bar"). Use group by and matplotlib functions.
season_group=day.groupby("season")[["casual",'registered']].sum().reset_index()
x=np.arange(len(season_group["season"]))
width=0.35
plt.bar(x-width/2,season_group["casual"],width,label="casual",color='skyblue')
plt.bar(x+width/2,season_group["registered"],width,label="registered",color='orange')
plt.xlabel('season')
plt.ylabel('number of users')
plt.title('Casual vs Registered users for each season')
plt.xticks(x,season_group["season"])
plt.legend()
plt.show()
```
