```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('WHO_time_series.csv',parse_dates=['Date_reported'])
```

```python
plt.plot(df['Date_reported'],df['Cumulative_cases'])              #line plot
plt.xlabel('Date_reported')
plt.ylabel('Cumulative_cases')
plt.title('Date vs Cumulative_cases')
plt.show()
```

```python
# df["Country"].unique()
china=df[df['Country']=='China']
India=df[df['Country']=='India']
```

```python
china.shape
plt.plot(china['Date_reported'],china['Cumulative_cases'],ls="--",color="red")              #line plot
plt.xlabel('Date_reported')
plt.ylabel('Cumulative_cases')
plt.title('Date vs Cumulative_cases for China')
plt.show()
plt.plot(India['Date_reported'],India['Cumulative_cases'],color="black")              #line plot
plt.xlabel('Date_reported')
plt.ylabel('Cumulative_cases')
plt.title('Date vs Cumulative_cases for India')
plt.show()
```

```python
plt.plot(china['Date_reported'],china['Cumulative_cases'],ls="--",color="red",label="China")              #line plot
plt.plot(India['Date_reported'],India['Cumulative_cases'],color="black",label="India")                    #line plot
plt.xlabel('Date_reported')
plt.ylabel('Cumulative_cases')
plt.title('Covid Cases')
plt.legend(loc="upper left")
plt.show()
```

```python
def Country_plot(country_name,linestyle,color):
    country_data=df[df['Country']==country_name]
    plt.plot(country_data['Date_reported'],country_data['Cumulative_cases'],ls=linestyle,color=color,label=country_name)              #line plot
    plt.xlabel('Date_reported')
    plt.ylabel('Cumulative_cases')
    plt.title('Covid Cases')
    plt.legend(loc="upper left")
    plt.show()
Country_plot('Italy','--','blue')
Country_plot('Spain','-.','green')
Country_plot('India',':','orange')
```

```python
def Country_graph(**kwargs):
    for country_name,style_color in kwargs.items():
        linestyle,color=style_color
        country_data=df[df['Country']==country_name]
        plt.plot(country_data['Date_reported'],country_data['Cumulative_cases'],ls=linestyle,color=color,label=country_name)              #line plot
    plt.xlabel('Date_reported')
    plt.ylabel('Cumulative_cases')
    plt.title('Covid Cases')
    plt.legend(loc="upper left")
    plt.show()
Country_graph(Italy=('--','blue'),Spain=('-.' ,'green'),India=(':','orange'),China=('--','red'))
```
